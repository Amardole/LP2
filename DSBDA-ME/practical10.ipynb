```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
```

```python
df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
```

```python
df
```

```python
# Add the target (species) column
df['species'] = iris_data.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
df.head()
```

```python
# Get column names and data types
print("Features and their data types:\n")
print(df.dtypes)

# You can also classify them manually
print("\nFeature Type Summary:")
print("Sepal length, Sepal width, Petal length, Petal width => Numeric")
print("Species => Nominal (categorical)")
```

```python

plt.hist(df['sepal length (cm)'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python

sns.boxplot(y=df['sepal length (cm)'], color='lightblue')
plt.title('Boxplot of Sepal Length')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
plt.hist(df['sepal width (cm)'], bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(y=df['sepal width (cm)'], color='violet')
plt.title('Boxplot of Sepal Width')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Plot histograms for numeric features
df.hist(figsize=(10, 6), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Create boxplots for all numeric features
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, y=column, x='species', palette='Set2')
    plt.title(f'Boxplot of {column} by Species')
plt.tight_layout()
plt.show()
```

```python
"""
Observations:

1. All four features (sepal length, sepal width, petal length, petal width) are numeric.
2. 'Species' is a nominal feature with three categories: setosa, versicolor, virginica.
3. Histograms show that petal length and width are strongly clustered for each species.
4. Boxplots clearly show that:
   - Setosa has shorter petals and sepals.
   - Virginica tends to have the largest features.
5. Outliers are visible in sepal width for Setosa and possibly a few values in other features.
"""
```
